```python
import numpy as np
import matplotlib.pyplot as plt

# 1. データの準備（-10 から 10 まで）
x = np.arange(-10, 10, 0.1)

# --- 実験：シグモイドを3段階で解剖する ---

# Step 1: exp(-x) の動きを見る
# これは「xが大きくなると急速に0になり、xがマイナスだと爆発的に増える」やつだ
step1 = np.exp(-x)

# Step 2: 分母 (1 + exp(-x)) の動きを見る
# さっきのやつに 1 を足しただけ。つまり「最低でも1以上」になる。
step2 = 1 + step1

# Step 3: 逆数 (1 / 分母) をとる
# 分母がデカければ 0 に近づき、分母が 1 に近ければ 1 になる。
y = 1 / step2

# --- 可視化（ここが大事！） ---
plt.figure(figsize=(10, 6))

# Step 1 は値がデカすぎるので、対数グラフにしないと見えないレベルだが、
# 今回はあえてそのまま表示して「暴れん坊」ぶりを確認する
plt.plot(x, step1, "--", label="Step 1: exp(-x)", color="green", alpha=0.5)

# Step 2
plt.plot(x, step2, "--", label="Step 2: 1 + exp(-x)", color="orange", alpha=0.5)

# Final Step (シグモイド)
plt.plot(x, y, linewidth=3, label="Final: Sigmoid", color="red")

plt.ylim(-1, 5)  # 見やすいように範囲を制限
plt.grid()
plt.legend()
plt.title("Anatomy of Sigmoid Function")
plt.show()
```

```python
# 実装のまとめ
def init_network():
    network = {}
    network["W1"] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network["b1"] = np.array([0.1, 0.2, 0.3])
    network["W2"] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network["b2"] = np.array([0.1, 0.2])
    network["W3"] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network["b3"] = np.array([0.1, 0.2])

    return network


def forward(network, x):
    W1, W2, W3 = network["W1"], network["W2"], network["W3"]
    b1, b2, b3 = network["b1"], network["b2"], network["b3"]

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)

    return y


# Shape of u
print(f"W1: {W1.shape}")
print(f"b1: {np.array([0.1, 0.2, 0.3]).shape}")
print(f"W2: {W2.shape}")
print(f"b2: {np.array([0.1, 0.2]).shape}")
print(f"W3: {W3.shape}")
print(f"b3: {np.array([0.1, 0.2]).shape}")

# Dot of u
print(f"a1: {np.dot(x, W1) + b1}")

network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)
print(y)
```

```python
import sys, os
import pickle
import numpy as np  # 【修正1】これがないと計算できない


# sigmoidとsoftmaxを定義（あるいは common.functions から import）
# ここでは教科書の流れに沿って定義を書いておく
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def softmax(x):
    c = np.max(x)
    exp_a = np.exp(x - c)  # オーバーフロー対策
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y


# パス設定（ここはさっきのままで良い）
sys.path.append(os.path.join(os.pardir, "deep-learning-from-scratch"))
from dataset.mnist import load_mnist

# ディレクトリ移動（必須）
os.chdir(r"D:\1BelltreeTechSub\DeepLearning-from-zero\deep-learning-from-scratch\ch03")


def get_data():
    # 【修正2 & 3】
    # flatten=True: 画像を1列（784要素）に平坦化しないと、行列の積(np.dot)ができない
    # normalize=True: 0~255を0.0~1.0に正規化する
    # one_hot_label=False: 正解ラベルは数字単体（例: 7）で受け取る。配列（例:[0,0...1...]）だと比較が面倒
    (x_train, t_train), (x_test, t_test) = load_mnist(
        normalize=True, flatten=True, one_hot_label=False
    )
    return x_test, t_test


def init_network():
    with open("sample_weight.pkl", "rb") as f:
        network = pickle.load(f)
    return network


def predict(network, x):
    W1, W2, W3 = network["W1"], network["W2"], network["W3"]
    b1, b2, b3 = network["b1"], network["b2"], network["b3"]

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y


# --- 実行ブロック ---

x, t = get_data()
network = init_network()

accuracy_cnt = 0
for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y)  # 最も確率が高い要素のインデックスを取得

    # one_hot_label=Falseにしたので、単純な数値比較が可能になる
    if p == t[i]:
        accuracy_cnt += 1

print("Accuracy:" + str(float(accuracy_cnt) / len(x)))

# お前がロードした「脳みそ」の解剖
print("\n--- Network Anatomy ---")
print("W1 shape:", network["W1"].shape)
print("W2 shape:", network["W2"].shape)
print("W3 shape:", network["W3"].shape)

# ある1枚の画像について、ネットワークがどう判断したか覗き見る
print("\n--- Insight of x[0] ---")
y_test = predict(network, x[0])
print("AIの予測確率分布:\n", np.round(y_test, 3))  # 小数点3桁で見やすく
print("AIの結論:", np.argmax(y_test))
print("正解:", t[0])
```
