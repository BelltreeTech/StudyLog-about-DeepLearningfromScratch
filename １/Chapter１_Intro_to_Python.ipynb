```python
# 関数
def hello():
    print("Hello World")


hello()


def hello(object):
    print("Hello" + object + "!")


hello("Cat")
```

```python
# クラス
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")

    def hello(self):
        print("Hello " + self.name + "!")

    def goodbye(self):
        print("goodbye " + self.name + "!")


m = Man("Mari")
m.hello()
m.goodbye()
```

```python
# Numpy配列の生成と算術計算
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])

print(x - y)
print(x * y)
print(x / y)
```

```python
# ブロードキャスト
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
print(A * B)

C = np.array([[1, 2], [3, 4]])
D = np.array([[10, 20], [30, 40]])
print(C * D)
```

```python
# 単純なグラフの描画
import numpy as np
import matplotlib.pyplot as plt

# データの作成
x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# pyplotの機能
import numpy as np
import matplotlib.pyplot as plt

# データの作成
x = np.arange(0, 6, 0.1)

y1 = np.sin(x)
y2 = np.cos(x)
# y3 = np.tan(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
# plt.plot(x, y3, label = "tan")
```
