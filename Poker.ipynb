```python
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank


class Deck:
    def __init__(self):
        self.cards = []
        for suit in range(4):
            for rank in range(2, 15):
                self.cards.append(Card(suit, rank))

    def shuffle(self):
        import random

        random.shuffle(Card(suit, rank))
```
